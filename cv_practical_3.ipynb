```python
from google.colab.patches import cv2_imshow 
import cv2
import numpy as np
import matplotlib.pyplot as plt
 
```

```python
image = cv2.imread('/content/drive/MyDrive/Computer_vision/Input images/prog_3.jpeg')
# cv2_imshow(image)

gray_scaled_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

path ='/content/drive/MyDrive/Computer_vision/Input images/g_prog_3.png'
# cv2_imshow(gray_scaled_img)

# cv2.imwrite(path, gray_scaled_img)

plt.rcParams["figure.figsize"]=(10,10)
plt.imshow(gray_scaled_img, cmap='gray')
plt.show()
```

```python
h_gram= np.asarray(gray_scaled_img)

flat = h_gram.flatten()

plt.hist(flat, bins=30)
```

```python
h_eq= cv2.equalizeHist(gray_scaled_img)
res = np.hstack((gray_scaled_img, h_eq))
h_res = np.asarray(res)
flat = h_res.flatten()
his = plt.hist(flat,bins=50)
plt.xlabel("pixel")
plt.ylabel("frequency")
plt.show()
cv2_imshow(res)
```
